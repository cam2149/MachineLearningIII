## 2.1. Flujo de Trabajo en Aprendizaje Supervisado

En aprendizaje supervisado, tenemos una variable objetivo conocida. El flujo de trabajo tÃ­pico en PyCaret es:

### ğŸŒŸ Flujo de Trabajo TÃ­pico en Aprendizaje Supervisado con PyCaret

1. **âš™ï¸ Setup**: Configura el entorno de trabajo, define el tipo de problema y la variable objetivo. AquÃ­ se hace el preprocesamiento de los datos: codificaciÃ³n de variables categÃ³ricas, manejo de valores nulos y divisiÃ³n en entrenamiento y prueba.

   ```python
   from pycaret.classification import *  # o pycaret.regression para regresiÃ³n
   clf = setup(data=df, target='variable_objetivo', session_id=123)
   ```

2. **ğŸ“ˆ Create Model**: Crea y entrena un modelo especÃ­fico o varios para comparaciÃ³n. PyCaret tiene una variedad de algoritmos supervisados listos para usar (por ejemplo, regresiÃ³n logÃ­stica, Ã¡rboles de decisiÃ³n, bosques aleatorios).

   ```python
   best_model = create_model('rf')  # 'rf' es el Random Forest como ejemplo
   ```

3. **ğŸ”§ Tune Model**: Ajusta los hiperparÃ¡metros del modelo para mejorar su precisiÃ³n y rendimiento. Este paso es crucial para optimizar el modelo y sacarle el mÃ¡ximo provecho.

   ```python
   tuned_model = tune_model(best_model)
   ```

4. **ğŸ“Š Plot Model**: Genera grÃ¡ficos para analizar el rendimiento del modelo. Puedes ver la matriz de confusiÃ³n, la curva ROC y mÃ¡s, para evaluar la eficacia de las predicciones del modelo.

   ```python
   plot_model(tuned_model, plot='confusion_matrix')
   ```

5. **ğŸ§ª Evaluate Model**: PyCaret ofrece una interfaz de evaluaciÃ³n completa donde puedes analizar distintas mÃ©tricas de rendimiento. Esto permite profundizar en el anÃ¡lisis y mejorar la comprensiÃ³n del modelo.

   ```python
   evaluate_model(tuned_model)
   ```

6. **ğŸ”® Predict Model**: Realiza predicciones en nuevos datos usando el modelo entrenado y optimizado.

   ```python
   predictions = predict_model(tuned_model, data=nuevos_datos)
   ```

7. **ğŸ’¾ Save Model**: Guarda el modelo para su uso posterior. Esto es Ãºtil para guardar el trabajo realizado y poder reutilizar el modelo sin necesidad de reentrenarlo.

   ```python
   save_model(tuned_model, 'mejor_modelo')
   ```

````python
# Importar las bibliotecas necesarias
from pycaret.classification import *
import pandas as pd

# Cargar el conjunto de datos
file_path = 'Accidentalidad.csv'  # Reemplaza con la ruta correcta
data = pd.read_csv(file_path)

# Preprocesamiento de datos
# Rellenar valores nulos y seleccionar las columnas relevantes
data['n_heridos'] = data['n_heridos'].fillna(0).astype(int)
data['n_muertos'] = data['n_muertos'].fillna(0).astype(int)

# Crear la columna de severidad en base a 'n_heridos' y 'n_muertos'
def calcular_severidad(row):
    if row['n_muertos'] > 0:
        return 'alto'
    elif row['n_heridos'] > 0:
        return 'medio'
    else:
        return 'bajo'

data['severidad'] = data.apply(calcular_severidad, axis=1)

# Seleccionar las columnas necesarias
data = data[['dia_semana_acc', 'hora_acc', 'condic_meteor', 'estado_super', 'clase_accidente', 'Causa_old', 'severidad']]

# ConfiguraciÃ³n de PyCaret
clf = setup(data=data, target='severidad', session_id=123,
            categorical_features=['dia_semana_acc', 'condic_meteor', 'estado_super', 'clase_accidente', 'Causa_old'])

# Comparar modelos y seleccionar el mejor
best_model = compare_models()

# Entrenar y afinar el mejor modelo
tuned_model = tune_model(best_model)
print("Modelo seleccionado:", tuned_model)

# Guardar el modelo entrenado para futuras predicciones
save_model(tuned_model, 'mejor_modelo_severidad_accidentes')

# FunciÃ³n para predecir la severidad del accidente con datos proporcionados por el usuario
def make_prediction(model):
    # Solicitar informaciÃ³n al usuario
    dia_semana_acc = input("Ingrese el dÃ­a de la semana (ej. lunes, martes): ")
    hora_acc = int(input("Ingrese la hora del accidente (0-23): "))
    condic_meteor = input("Ingrese las condiciones meteorolÃ³gicas (ej. soleado, lluvioso): ")
    estado_super = input("Ingrese el estado de la superficie de la vÃ­a (ej. seca, mojada): ")
    clase_accidente = input("Ingrese la clase de accidente (ej. choque, atropello): ")
    causa_old = input("Ingrese la causa probable del accidente (ej. impericia, embriaguez): ")

    # Crear un DataFrame con los datos ingresados
    new_data = pd.DataFrame({
        'dia_semana_acc': [dia_semana_acc],
        'hora_acc': [hora_acc],
        'condic_meteor': [condic_meteor],
        'estado_super': [estado_super],
        'clase_accidente': [clase_accidente],
        'Causa_old': [causa_old]
    })

    # Realizar la predicciÃ³n
    prediction = predict_model(model, data=new_data)
    severidad_predicha = prediction['Label'][0]  # Obtener la predicciÃ³n de severidad
    print(f"La severidad predicha del accidente es: {severidad_predicha}")

# Cargar el modelo guardado para futuras predicciones
mejor_modelo = load_model('mejor_modelo_severidad_accidentes')

# Realizar una predicciÃ³n interactiva con el usuario
make_prediction(mejor_modelo)
````

## 2.2. Flujo de Trabajo en Aprendizaje No Supervisado

En aprendizaje no supervisado, no tenemos una variable objetivo definida y buscamos patrones dentro de los datos, como clusters o anomalÃ­as.
Un flujo de trabajo tÃ­pico en el mÃ³dulo no supervisado de PyCaret consiste en seguir estos 6 pasos en este orden:

**Setup** â¡ï¸ **Create Model** â¡ï¸ **Assign Labels** â¡ï¸ **Analyze Model** â¡ï¸ **Prediction** â¡ï¸ **Save Model**

1. **Setup**: InicializaciÃ³n del entorno para clustering o detecciÃ³n de anomalÃ­as.
   ```python
   from pycaret.clustering import *
   clf = setup(data=df)
   ```

2. **Create Model**: CreaciÃ³n y entrenamiento de un modelo de clustering o detecciÃ³n de anomalÃ­as.
   ```python
   kmeans = create_model('kmeans')  # Por ejemplo, un modelo de K-Means
   ```

3. **Assign Labels**: AsignaciÃ³n de etiquetas a los datos basadas en los clusters identificados.
   ```python
   labeled_data = assign_model(kmeans)
   ```

4. **Plot Model**: VisualizaciÃ³n de los clusters o resultados de detecciÃ³n de anomalÃ­as.
   ```python
   plot_model(kmeans, plot='cluster')
   ```

5. **Evaluate Model**: EvaluaciÃ³n y anÃ¡lisis de los resultados del modelo.
   ```python
   evaluate_model(kmeans)
   ```

6. **Save Model**: Guardado del modelo de clustering para futuros anÃ¡lisis.
   ```python
   save_model(kmeans, 'kmeans_clustering_model')
   ```

Estos flujos de trabajo en PyCaret permiten una implementaciÃ³n rÃ¡pida de experimentos de machine learning supervisado y no supervisado, haciendo que la validaciÃ³n y comparaciÃ³n de modelos sea mÃ¡s eficiente.

Este ejemplo prÃ¡ctico utiliza PyCaret para realizar un anÃ¡lisis no supervisado de detecciÃ³n de anomalÃ­as, identificando datos atÃ­picos en un conjunto de datos. Este tipo de anÃ¡lisis es Ãºtil en aplicaciones como la detecciÃ³n de fraudes, mantenimiento predictivo y monitoreo de calidad.

## Paso 1: Cargar LibrerÃ­as y el Dataset ğŸ“¥

Primero, cargamos las librerÃ­as necesarias y el conjunto de datos para el anÃ¡lisis.

```python
import pandas as pd
from pycaret.anomaly import *

# Cargar el dataset de ejemplo

data.head()
```

## Paso 2: Configurar el Entorno de PyCaret para AnomalÃ­as ğŸ› ï¸

Usamos la funciÃ³n `setup()` de PyCaret para configurar el entorno. Dado que es un anÃ¡lisis no supervisado, no necesitamos especificar una variable objetivo.

```python
# Configurar PyCaret para la detecciÃ³n de anomalÃ­as
setup(data=data, session_id=123)
```

> **ğŸ”µ Nota**: `session_id` garantiza la reproducibilidad, y `setup()` prepara los datos para el anÃ¡lisis.

## Paso 3: Crear el Modelo de DetecciÃ³n de AnomalÃ­as ğŸ†

Creamos un modelo de detecciÃ³n de anomalÃ­as. En este ejemplo, usamos el modelo `IForest` (Isolation Forest), que es popular en la detecciÃ³n de anomalÃ­as.

```python
# Crear un modelo de detecciÃ³n de anomalÃ­as (Isolation Forest)

```

> **ğŸŸ¢ Consejo**: `IForest` es efectivo para detectar datos atÃ­picos en grandes conjuntos de datos, basado en el aislamiento de puntos inusuales.

## Paso 4: Ajustar el Modelo ğŸ”§

Si se desea, podemos optimizar algunos parÃ¡metros del modelo para mejorar su precisiÃ³n en la detecciÃ³n de anomalÃ­as.

```python
# OptimizaciÃ³n de parÃ¡metros (si es necesario)
tuned_model = tune_model(model)
```

> **ğŸŸ¡ Importante**: La optimizaciÃ³n mejora el rendimiento del modelo, pero no siempre es necesaria en detecciÃ³n de anomalÃ­as.

## Paso 5: Etiquetar las AnomalÃ­as ğŸ“Š

Usamos el modelo ajustado para etiquetar las observaciones en el conjunto de datos. Las observaciones se etiquetarÃ¡n como "1" si se consideran anÃ³malas, o "0" si son normales.

```python
# Etiquetar anomalÃ­as en el conjunto de datos original

anomaly_labels.head()
```

> **ğŸ”´ Alerta**: Estas etiquetas pueden ser usadas para analizar o filtrar solo los datos que se consideran anÃ³malos.

## Paso 6: Evaluar el Modelo de AnomalÃ­as ğŸ“ˆ

Evaluamos el modelo utilizando grÃ¡ficos que muestren la distribuciÃ³n y localizaciÃ³n de las anomalÃ­as.


Extraemos los datos etiquetados como anomalÃ­as para su anÃ¡lisis posterior.

> **ğŸ”µ Resultado**: Este paso muestra los datos considerados anÃ³malos, permitiendo un anÃ¡lisis detallado de estos casos.

````python


# Importar PyCaret para clustering, pandas para manipulaciÃ³n de datos
from pycaret.clustering import *
import pandas as pd

# Cargar el conjunto de datos de enfermedades cardÃ­acas
from pycaret.datasets import get_data
data = get_data('heart_disease')

# Configurar PyCaret para clustering
cluster_setup = setup(data, session_id=123, normalize=True, ignore_features=['Disease'])

# Crear el modelo de clustering KMeans
kmeans_model = create_model('kmeans', num_clusters=3)  # Ajusta el nÃºmero de clusters segÃºn sea necesario

# FunciÃ³n para solicitar al usuario sus datos de salud y predecir su cluster
def predict_cluster(model):
    print("Por favor, ingresa los siguientes datos de salud:")

    # Solicitar datos del usuario
    age = int(input("Edad: "))
    sex = int(input("Sexo (1=Masculino, 0=Femenino): "))
    cp = int(input("Tipo de dolor en el pecho (0-3): "))
    trestbps = int(input("PresiÃ³n arterial en reposo (mm Hg): "))
    chol = int(input("Colesterol sÃ©rico (mg/dl): "))
    fbs = int(input("AzÃºcar en sangre en ayunas (1=Si, 0=No): "))
    restecg = int(input("Resultados del electrocardiograma en reposo (0-2): "))
    thalach = int(input("Frecuencia cardÃ­aca mÃ¡xima alcanzada: "))
    exang = int(input("Â¿Angina inducida por el ejercicio? (1=SÃ­, 0=No): "))
    oldpeak = float(input("DepresiÃ³n del ST: "))
    slope = int(input("Pendiente del ST (0-2): "))
    ca = int(input("NÃºmero de vasos coloreados por fluoroscopia (0-3): "))
    thal = int(input("Tipo de talasemia (1-3): "))

    # Crear un DataFrame con los nombres de columna correctos
    user_data = pd.DataFrame({
        'age': [age],
        'sex': [sex],
        'chest pain type': [cp],
        'resting blood pressure': [trestbps],
        'serum cholestoral in mg/dl      ': [chol],
        'fasting blood sugar > 120 mg/dl ': [fbs],
        'resting electrocardiographic results': [restecg],
        'maximum heart rate achieved  ': [thalach],
        'exercise induced angina    ': [exang],
        'oldpeak ': [oldpeak],
        'slope of peak': [slope],
        'number of major vessels ': [ca],
        'thal': [thal]
    })

    # Asignar al usuario a un cluster
    prediction = predict_model(model, data=user_data)
    user_cluster = prediction['Cluster'][0]
    print(f"\nEl usuario pertenece al cluster: {user_cluster}")

    return user_cluster

# Realizar la predicciÃ³n de cluster
usuario_cluster = predict_cluster(kmeans_model)

# AnÃ¡lisis de clusters para determinar niveles de salud
clustered_data = assign_model(kmeans_model)
cluster_health = clustered_data.groupby('Cluster').mean()

# Mostrar resultados de salud por cluster para comparar perfiles de riesgo
print("\nPromedio de caracterÃ­sticas de salud por cluster (valores mÃ¡s bajos en colesterol, presiÃ³n, etc. pueden indicar mejor salud):")
print(cluster_health)

# EvaluaciÃ³n de salud por cluster
print("\nInterpretaciÃ³n general:")
for cluster, stats in cluster_health.iterrows():
    if stats['serum cholestoral in mg/dl      '] < clustered_data['serum cholestoral in mg/dl      '].mean() and stats['resting blood pressure'] < clustered_data['resting blood pressure'].mean():
        print(f"Cluster {cluster} tiende a representar perfiles mÃ¡s saludables (colesterol y presiÃ³n arterial mÃ¡s bajos).")
    else:
        print(f"Cluster {cluster} tiende a representar perfiles menos saludables (valores mÃ¡s altos en colesterol o presiÃ³n arterial).")
````
